```python
import requests
import os
from PyPDF2 import PdfReader, PdfWriter
from fpdf import FPDF  # 텍스트를 PDF 페이지에 추가하기 위한 라이브러리
import re
# API key 설정
api_key = "up_gxdEHQGG3mSRCtJhbWvQ37jLInELU"

# 폴더 내의 PDF 파일 경로를 지정
folder_path = "/Users/minseok/Desktop/ragnlp/streamlit/partial_data/"

# OCR API를 호출하고 텍스트를 가져오는 함수
def extract_text_from_ocr(filename):
    url = "https://api.upstage.ai/v1/document-ai/ocr"
    headers = {"Authorization": f"Bearer {api_key}"}
    
    # PDF 파일을 읽고 API 요청을 보냄
    with open(filename, "rb") as file:
        files = {"document": file}
        response = requests.post(url, headers=headers, files=files)
    
    # OCR API의 응답에서 텍스트 추출
    if response.status_code == 200:
        return response.json().get('text', '')
    else:
        print(f"OCR 실패: {filename} 응답 코드 {response.status_code}")
        return None

from tqdm import tqdm
import re

for i in tqdm(range(8, 26), desc="Processing files"):
    filename = f'/Users/minseok/Desktop/ragnlp/streamlit/partial_data/2024_graduation_file-part-{i}.pdf'
    print(f"처리 중: {filename}")

    # 1. OCR을 통해 텍스트 추출
    ocr_text = extract_text_from_ocr(filename)

    # 2. OCR 텍스트가 있으면 새로운 PDF에 저장
    if ocr_text:
        print(len(ocr_text))
        text = ocr_text   
        
        # 전처리: 특정 패턴을 기반으로 줄바꿈 및 공백 추가
        
        # 불필요한 기호와 공백 제거
        text = re.sub(r'■|s+', ' ', text)
        
        # 괄호 안의 불필요한 공백 제거
        text = re.sub(r'\(\s+', '(', text)
        text = re.sub(r'\s+\)', ')', text)
        
        # 특정 불필요한 문구 제거
        text = re.sub(r'공동체 정신을 지닌 혁신적 리더', '', text)
        text = re.sub(r'진리와 자유를 향한 연세의 도전\nExcellence, Innovation, Engagement\n', '', text)
        text = re.sub(r'연세대학교 대학요람2024', '', text)
        
        # 불필요한 인용 기호 제거
        text = re.sub(r'[‘’\'\']', '', text)
        text = re.sub(r',', '', text)

        # 처리된 텍스트를 리스트에 저장
        pages[i-1].page_content = text
    else:
        print(f"OCR 실패: {filename}")
```

```python
text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=5000,
    chunk_overlap=500,
    encoding_name='cl100k_base'
)
```

```python
documents = text_splitter.split_documents(pages2)
documents2 = documents
len(documents)
```

```python
print(documents[200])
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings_model = HuggingFaceEmbeddings(
    model_name='jhgan/ko-sroberta-nli',
    model_kwargs={'device':'cpu'},
    encode_kwargs={'normalize_embeddings':True},
)
```

```python
from langchain_community.vectorstores import FAISS
from langchain_community.vectorstores.utils import DistanceStrategy

vectorstore2 = FAISS.from_documents(documents2,
                                   embedding = embeddings_model,
                                   distance_strategy = DistanceStrategy.COSINE
                                  )
```
